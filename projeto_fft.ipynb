```
import scipy.io
from scipy.io import wavfile
from scipy.fft import fft, ifft
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output, Audio

pontos = 150000

freq, data = scipy.io.wavfile.read("/content/datasets_453_923_set_a_Aunlabelledtest__201101051105.wav")
samples = data.shape[0]
T = 1/freq
x = np.linspace(0,samples*T,samples)
fourier = np.fft.fft(data)
freq1 = np.fft.fftfreq(samples,T)

half_list = len(freq1)//2
freq2 = freq1[:half_list]

fourier = fourier/max(fourier)
fourier2 = abs(fourier[:half_list])

fourier[-pontos:] =0

fi = np.fft.ifft(fourier)


scipy.io.wavfile.write("/content/filtered_audio1.wav", freq, fi.astype(data.dtype))


#for plotting the graphics
plt.plot(freq1,fourier)
plt.grid()
plt.show()

plt.plot(freq1,fourier)
plt.grid()
plt.show()

plt.plot(data)
plt.grid()
plt.show()

plt.plot(fi)
plt.grid()
plt.show()

```
